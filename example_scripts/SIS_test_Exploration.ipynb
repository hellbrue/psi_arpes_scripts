```python
# Load EDC example data

from arpes.io import load_data
f = load_data('EDC_AN_P2p5_25eV_CIRCP_LT_10eVpass_0001.h5', location="ULTRA")
```

```python
f.S.plot()
```

```python
f.spectrumf.sel(eV=slice(-0.2, 0)).S.plot()
```

```python
f.sel(eV=slice(-0.2, 0)).S.plot()
```

```python
# get the EDC nearly at the gamma point
f.sel(phi=-9.5, method="nearest").S.plot()
```

```python
# get the EDC nearly at the gamma point
f.sel(phi=slice(-10, -9)).mean("phi").S.plot()
f.sel(phi=slice(-10, -9)).sum("phi").S.plot()
```

```python
f.isel(phi=550).S.plot()
```

```python
f.isel(phi=slice(None, 550)).S.plot()
```

```python
f.mean("eV", keep_attrs=True).S.plot()
```

```python
# to get the value as a plain float instead of a scalar dataset
# use `.mean(["eV", "phi"]).item()` instead
f.spectrum.mean(["eV", "phi"], keep_attrs=True)
```
